## Signals and frequency components

```python
%pylab inline
%config InlineBackend.figure_format='svg'
from IPython.display import Image
```

```python
x = linspace(0, 1, 101) # 101 samples from 0 to 1 (seconds)
s1 = sin(2*pi*x) # 1 Hz
s5 = sin(5*2*pi*x) # 1 Hz
s40 = sin(25*2*pi*x) # 1 Hz
plot(x,s1)
plot(x,s5)
plot(x,s40)
plot(x,s1, '.')
```

```python
combined_signal = 0.3 * (s1+s5+s40)
plot(combined_signal)
```

## Convolution and Filter Kernels

```python
kernel_avg = zeros(101)
for i in range(33,66):
    kernel_avg[i] = 1.0 / 34
plot(kernel_avg, ".")
```

```python
filtered_signal = convolve(combined_signal, kernel_avg, mode='same')
```

```python
plot(filtered_signal)
```

## Discrete / Fast Fourier Transform

```python
Image("img/the_fourier_transform_explained_in_one_sentence.png")
# http://blog.revolutionanalytics.com/2014/01/the-fourier-transform-explained-in-one-sentence.html
```

```python
def plotSpectrum(y,Fs):
 """
 Plots a Single-Sided Amplitude Spectrum of y(t)
 http://glowingpython.blogspot.de/2011/08/how-to-plot-frequency-spectrum-with.html
 """
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 #frq = frq[0:int(n/2)] # one side frequency range
 Y = fft.fft(y)/n # fft computing and normalization
 #Y = Y[0:int(n/2)]

 plot(frq,abs(Y),'r') # plotting the spectrum
 xlabel('Frequency (Hz)')
 ylabel('Intensity')
```

```python
Fs = 100 # sampling rate of x
subplot(2,1,1)
plot(x,combined_signal)
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum(combined_signal,Fs)
show()
```
