# Reaction Time Experiment

Short experiment description

```python
%matplotlib inline
from pylab import *
```

Import csv data

```python
datatype = [('AGE', '<i8'), ('EXPECTED_BUTTON', 'S1'), ('GENDER', 'S6'), ('HANDEDNESS', 'S5'), 
            ('HAS_DISTRACTION', '?'), ('ORDER', 'S11'), ('PARTICIPANT', '<i8'), ('PRESSED_KEY', 'S1'), 
            ('REACTION_TIME', '<f8'), ('REPETITIONS', '<i8'), ('SIGNAL_MODE', 'S13'), ('TIMESTAMP', '<f8'), 
            ('TIME_BETWEEN_SIGNALS_MS', '<i8'), ('WAS_CORRECT_BUTTON', '?')]
data1 = genfromtxt("Test_Result1.csv", dtype=datatype, delimiter=',', names=True)
data2 = genfromtxt("Test_Result2.csv", dtype=datatype, delimiter=',', names=True)
data3 = genfromtxt("Test_Result3.csv", dtype=datatype, delimiter=',', names=True)
data4 = genfromtxt("Test_Result4.csv", dtype=datatype, delimiter=',', names=True)
data = concatenate((data1, data2, data3, data4), axis=0)
```

Extract relevant data

```python
pre_attentive = array([])
attentive = array([])
pre_attentive_distr = array([])
attentive_distr = array([])

for d in data:
    if d['SIGNAL_MODE'] == b'pre-attentive' and d['HAS_DISTRACTION'] == False:
        pre_attentive = append(pre_attentive, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'attentive' and d['HAS_DISTRACTION'] == False:
        attentive = append(attentive, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'pre-attentive' and d['HAS_DISTRACTION'] == True:
        pre_attentive_distr = append(pre_attentive_distr, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'attentive' and d['HAS_DISTRACTION'] == True:
        attentive_distr = append(attentive_distr, d['REACTION_TIME'])
```

Draw Scatterplot

```python
labels = ["pre-attentive", "attentive", "pre-attentive distracted", "attentive distracted"]
x_axis = [1,2,3,4]

scatter([1]*len(pre_attentive), pre_attentive, s=100, alpha=0.1)
scatter([2]*len(attentive), attentive, s=100, alpha=0.1)
scatter([3]*len(pre_attentive_distr), pre_attentive_distr, s=100, alpha=0.1)
scatter([4]*len(attentive_distr), attentive_distr, s=100, alpha=0.1)
xticks(x_axis, labels, rotation="vertical")
```
