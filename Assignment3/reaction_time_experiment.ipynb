# Reaction Time Experiment

In this experiment the reaction times of a testperson were measured, while executing some specific tasks. There were four different setups of the experiment which had to be done by all users just in a different order. The users had to press the "v" button, when they saw one rectangle or an even number on the screen. The users had to press the "b" button when they saw two rectangles or an odd number on the screen. The experiment was conducted four times per test person. One time they had to distinguish between one and two rectangels with and without distraction. And the other time the test persons had to distinguish between an even and odd number with and without distraction. Each key press was logged in a csv, and the reaction times according to this experiment setup will be evaluated int the following.

```python
from pylab import *
import matplotlib.patches as m_patches
import scipy.stats as stats
%matplotlib inline
```

Import csv data

```python
datatype = [('AGE', '<i8'), ('EXPECTED_BUTTON', 'S1'), ('GENDER', 'S6'), ('HANDEDNESS', 'S5'), 
            ('HAS_DISTRACTION', '?'), ('ORDER', 'S11'), ('PARTICIPANT', '<i8'), ('PRESSED_KEY', 'S1'), 
            ('REACTION_TIME', '<f8'), ('REPETITIONS', '<i8'), ('SIGNAL_MODE', 'S13'), ('TIMESTAMP', '<f8'), 
            ('TIME_BETWEEN_SIGNALS_MS', '<i8'), ('WAS_CORRECT_BUTTON', '?')]
data1 = genfromtxt("Test_Result1.csv", dtype=datatype, delimiter=',', names=True)
data2 = genfromtxt("Test_Result2.csv", dtype=datatype, delimiter=',', names=True)
data3 = genfromtxt("Test_Result3.csv", dtype=datatype, delimiter=',', names=True)
data4 = genfromtxt("Test_Result4.csv", dtype=datatype, delimiter=',', names=True)
data = concatenate((data1, data2, data3, data4), axis=0)
```

Extract relevant data

```python
pre_attentive = array([])
pre_attentive_false = array([])
attentive = array([])
attentive_false = array([])
pre_attentive_distr = array([])
pre_attentive_distr_false = array([])
attentive_distr = array([])
attentive_distr_false = array([])

for d in data:
    if d['SIGNAL_MODE'] == b'pre-attentive' and d['HAS_DISTRACTION'] == False:
        if d['WAS_CORRECT_BUTTON'] == False:
            pre_attentive_false = append(pre_attentive_false, d['REACTION_TIME'])
        else:
            pre_attentive = append(pre_attentive, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'attentive' and d['HAS_DISTRACTION'] == False:
        if d['WAS_CORRECT_BUTTON'] == False:
            attentive_false = append(attentive_false, d['REACTION_TIME'])
        else:
            attentive = append(attentive, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'pre-attentive' and d['HAS_DISTRACTION'] == True:
        if d['WAS_CORRECT_BUTTON'] == False:
            pre_attentive_distr_false = append(pre_attentive_distr_false, d['REACTION_TIME'])
        else:
            pre_attentive_distr = append(pre_attentive_distr, d['REACTION_TIME'])
    elif d['SIGNAL_MODE'] == b'attentive' and d['HAS_DISTRACTION'] == True:
        if d['WAS_CORRECT_BUTTON'] == False:
            attentive_distr_false = append(attentive_distr_false, d['REACTION_TIME'])
        else:
            attentive_distr = append(attentive_distr, d['REACTION_TIME'])
```

Shuffle method for spreading the scatter plot points for better readability

```python
def shuffle_x(l, amount=0.5):
    from random import random
    shuffled = []
    for val in l:
        shuffled.append(val + amount * (random() - 0.5))
    return shuffled
```

In the next code snippets the scatter plots will be drawn

Setup the legend, x and y axis description for the plot.
Also the plot itself will be drawn in the next cell.

```python
#setup legend, x-axis, y-axis
blue_patch = m_patches.Patch(color = "#00008b", label = 'right button clicked', alpha = 0.5)
red_patch = m_patches.Patch(color = "red", label = "false button clicked", alpha = 0.5)
legend(handles = [blue_patch, red_patch], loc='lower left', bbox_to_anchor=(1, 0.5))

labels = ["pre-attentive", "attentive", "pre-attentive distracted", "attentive distracted"]
x_axis = [1,2,3,4]
xticks(x_axis, labels, rotation = 40)

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
       }
ylabel('time (min)', fontdict=font)
subplot().set_ylim(0, 1.0)

#scatter the points in the plot
scatter(shuffle_x([1]*len(pre_attentive)), pre_attentive, s=100, alpha=0.1, c = "#00008b")
scatter(shuffle_x([1]*len(pre_attentive_false)), pre_attentive_false, s=100, alpha=0.5, c = "red")
scatter(shuffle_x([2]*len(attentive)), attentive, s=100, alpha=0.1, c= "#00008b")
scatter(shuffle_x([2]*len(attentive_false)), attentive_false, s=100, alpha=0.5, c = "red")
scatter(shuffle_x([3]*len(pre_attentive_distr)), pre_attentive_distr, s=100, alpha=0.1, c = "#00008b")
scatter(shuffle_x([3]*len(pre_attentive_distr_false)), pre_attentive_distr_false, s=100, alpha=0.5, c = "red")
scatter(shuffle_x([4]*len(attentive_distr)), attentive_distr, s=100, alpha=0.1, c="#00008b")
scatter(shuffle_x([4]*len(attentive_distr_false)), attentive_distr_false, s=100, alpha=0.5, c = "red")
```

Setup data arrays for next plots. The arrays for false and true button clicks are merged back together.

```python
pre_attentive = concatenate((pre_attentive, pre_attentive_false), axis = 0)
attentive = concatenate((attentive, attentive_false), axis = 0)
pre_attentive_distr = concatenate((pre_attentive_distr, pre_attentive_distr_false), axis = 0)
attentive_distr = concatenate((attentive_distr, attentive_distr_false), axis = 0)
```

```python
#setup legend, x-axis, y-axis
green_patch = m_patches.Patch(color = "#7fff00", label = 'pre-attentive', alpha = 0.5)
brown_patch = m_patches.Patch(color = "#b8860b", label = 'attentive', alpha = 0.5)
blue_patch = m_patches.Patch(color = "#00008b", label = 'pre-attentive distracted', alpha = 0.5)
violett_patch = m_patches.Patch(color = "#9932cc", label = 'attentive distracted', alpha = 0.5)
legend(handles = [green_patch, brown_patch, blue_patch, violett_patch], loc='lower left', bbox_to_anchor=(1, 0.5))

scatter([1]*len(pre_attentive), pre_attentive, s=100, alpha=0.1, c = "#7fff00")
scatter([1]*len(attentive), attentive, s=100, alpha=0.1, c= "#b8860b")
scatter([1]*len(pre_attentive_distr), pre_attentive_distr, s=100, alpha=0.1, c = "#00008b")
scatter([1]*len(attentive_distr), attentive_distr, s=100, alpha=0.1, c="#9932cc")
```

Draw the bar plots

```python
bar(0, mean(pre_attentive), yerr=std(pre_attentive))
bar(1, mean(attentive), yerr=std(attentive))
bar(2, mean(pre_attentive_distr), yerr=std(pre_attentive_distr))
bar(3, mean(attentive_distr), yerr=std(attentive_distr))
```

Format new arrays for the t-tests:

```python
attentive_reaction_times = concatenate((attentive, attentive_distr), axis = 0)
pre_attentive_reaction_times = concatenate((pre_attentive, pre_attentive_distr), axis = 0)
undistracted_reaction_times = concatenate((pre_attentive, attentive), axis = 0)
distracted_reaction_times = concatenate((pre_attentive_distr, attentive_distr), axis = 0)
```

T-Test: compare attentive and pre-attentive, compare distracted and undistracted

```python
stats.ttest_ind(attentive_reaction_times, pre_attentive_reaction_times, equal_var = False)
```

```python
stats.ttest_ind(undistracted_reaction_times, distracted_reaction_times, equal_var= False)
```
