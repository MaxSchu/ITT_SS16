# Filtering: Moving Average Filter

```python
rapid_change = zeros(101)
rapid_change[40:69] = 1
plot(rapid_change)
```

```python
filtered_signal = []
num_samples = 20
for i in range(len(rapid_change)):
    filtered_signal.append(sum(rapid_change[i:i+num_samples]/num_samples))
plot(filtered_signal)
```

```python
# weighted, adaptive moving average
```

# Convolution and Filter Kernels

```python
kernel_avg = zeros(101)
for i in range(47, 53):
    kernel_avg[i] = 1.0 / 6
plot(kernel_avg, ".")
```

```python
plot(signal)
```

```python
plot(noisy_signal)
```

```python
filtered_signal = convolve(noisy_signal, kernel_avg, mode='same')
plot(filtered_signal)
```

```python
kernel_inv = zeros(101)
kernel_inv = -1
plot(kernel_avg, ".")
```

```python
inverted_signal = convolve(signal, kernel_inv, mode = "same")
plot(inverted_signal)
plot(signal)
```

```python
kernel_drv = zeros(101)
kernel_drv[49] = 1
kernel_drv[50] = -1
plot(kernel_drv, ".")
```

```python
devirate = convolve(signal, kernel_drv, mode = "same")
plot(signal)
plot(devirate)
```
